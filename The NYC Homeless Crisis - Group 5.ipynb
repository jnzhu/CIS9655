## Subway Complaints

This chart shows the number of complaints from September 2018 to February 2020. There was a significant increase in the number of complaints in the winter of 2019 and 2020. For instance, there were 550 complaints in February 2020, which is more than twice the amount of complaints in February 2019. 

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
import matplotlib.dates as md
df = pd.read_csv('subway_complaints.csv', index_col=0, parse_dates=True)
df.head(20)
```

```python
plt.figure(figsize = (10,8))
sns.set_style("white")
p = sns.lineplot(data = df)
p.set_xlabel("Date", fontsize = 15)
p.set_ylabel("Number of Complaints", fontsize = 15)
p.set_title("Subway and Bus Complaints Related to Homelessness", fontsize= 20, y= 1.03)
```

## Unsheltered Homeless in the Subways

This graph indicates the number of unsheltered individuals located in the subway based on the HOPE count each year. HOPE is the homeless outreach population estimate and it’s conducted every January. There’s a decrease in 2020 and 2021, mainly because they had to make some changes to the survey process due to the pandemic. 

```python
subway = pd.read_csv('HOPE survey unsheltered areas (unaltered).csv',index_col=0, parse_dates=True)
subway.head(20)
```

```python
subway_filt = subway.loc[subway['AREA'] == 'Subways']
subway_filt
```

```python
plt.figure(figsize = (15,8))
sns.set_style("white")
p = sns.lineplot(data = subway_filt, markers=True)
p.set_xlabel("Year", fontsize = 15)
p.set_ylabel("HOPE Estimate", fontsize = 15)
p.set_title("Unsheltered Individuals in the Subway", fontsize= 20, y=1.03)
```

# How has NYC responded to the homeless crisis thus far? 

## Sheltered Individuals

The number of sheltered individuals approximately remained in the 50,000 to 60,000 bracket from 2014 to 2020. Beginning in 2020 when COVID-19 occurred, this number has dropped to nearly 40,000 in the current day. During the pandemic, the city transitioned homeless individuals from shelters to hotels in an effort to curb COVID-19's spread. The city Although there are fewer individuals now staying in shelters, the Department of Homeless Services (hereafter referred to as the "DHS") retained its shelter operations budget at an all-time high of nearly 810 million dollars. 

```python
df = pd.read_csv("DHS_Daily_ReportPython.csv")
df.head()
```

```python
df["Date of Census"] = pd.to_datetime(df["Date of Census"], format = "%Y-%m-%d")
fig,ax=plt.subplots()
sns.lineplot(ax = ax, x="Date of Census",y="Total Individuals in Shelter", data=df).set_title('Total Individuals in Shelters 2014-2022')
plt.yticks([0,10000,20000,30000,40000,50000,60000,70000,80000,90000])
plt.show()
```

## Unsheltered Individuals

The DHS drastically reduced the homeless Prevention & Aftercare budget beginning in 2017. In 2016 and 2017, the allocated budget was respectively about 65 million dollars and 17.5 million dollars. This budget reached new lows in 2019 with 57,000 dollars and in 2021 with 90,000 dollars. When the budget was at a record high in 2016, unsheltered individuals decreased by more than 1,000. After budget cuts were implemented, this number saw a steady rise once again. A point of note the DHS should consider when looking at the rise of unsheltered individuals is COVID-19 in 2020.  

```python
total = pd.read_csv("Hope survey unsheltered total.csv")
total.head()
fig,ax=plt.subplots()
x = sns.lineplot(ax = ax, x="YEAR",y="PEOPLE",data=total,markers=True,dashes=False)
x.set_title("Total Unsheltered Individuals", y=1.03)
x.set(yticks=np.arange(0,3250,500))
x.set_yticklabels(np.arange(0,3250,500))
ax.axvline(x=2016, color='g', label='Prevention Budget at Highest', linestyle='--', linewidth=2)
ax.axvline(x=2017, color='m', label='Prevention Budget Cuts', linestyle='--', linewidth=2)
ax.legend()
plt.show()
```

Since 2017, NYC has released a breakdown of total unsheltered individuals by area. However, the city elected to not publish this data completed in 2021. Only the area that saw decreases had their numbers published, and there was no mention of increases although the total number of unsheltered inviduals saw an increase into 2021. The city should be more forthcoming with this data and focus more on the prevention and aftercare of the homeless population

```python
hope = pd.read_csv("Hope survey unsheltered areas.csv")
hope.head()
```

```python
g = sns.relplot(x="YEAR",y="PEOPLE",data=hope, kind="line",style="AREA",hue="AREA",markers=True,dashes=False)
g.fig.suptitle("Unsheltered Individuals by Area", y=1.03)
plt.show()
```
